```python
import os 
from Bio import Phylo
import numpy as np
import math
```

```python
def calculateDensity(tree, c, K, N, log=False):
    # get all events
    events = []

    depths = tree.depths() # root has depth of 0
    max_depth= max(depths.values())
    inner = tree.get_nonterminals()
    outer = tree.get_terminals()

    for n in inner:
        height = max_depth - depths[n]
        if len(n.clades) == 2: # coalescent
            events.append(("COALESCENT", height))
        elif len(n.clades) == 1 and "active" in n.comment: # transition
            events.append(("TRANSITION D-A", height))
        elif len(n.clades) == 1 and "dormant" in n.comment: # transition
            events.append(("TRANSITION A-D", height))
        else:
            print(len(n.clades), n.comment)
            raise Exception
    
    for n in outer:
        height = max_depth - depths[n]
        # if height <= 0.00001:
            # height = 0
        if "active" in n.comment:
            events.append(("SAMPLE A", height))
        elif "dormant" in n.comment:
            events.append(("SAMPLE D", height))
        else:
            raise Exception
    
    events.sort(key=lambda x: x[1])
    # print(events)
    # print(list(map(lambda x: x[1], events)))

    num_active, num_dormant, last_time, logP = 0, 0, 0, 0
    for event, time in events:
        # print(event, time)
        theta = N*2
        if time >= last_time: 
            # add time contribution
            rate = ((1/theta) * math.comb(num_active, 2)) + (num_active * c) + (num_dormant * c * K)
            logP -= rate * (time - last_time)
            if log:
                print("TIME:", (time - last_time))
                print("TIME CONTRIBUTION:", rate * (time - last_time))
            last_time = time
        else:
            print(time, last_time)
            raise Exception

        # event contribution
        if event == 'COALESCENT':
            logP += np.log(1/theta)
            if log:
                print("COALESCENT: ", np.log(1/theta))
            num_active -= 1

        elif event == 'TRANSITION D-A':
            logP += np.log(c*K)
            num_active += 1
            num_dormant -= 1
            if log:
                print("TRANSITION D-A: ", np.log(c*K))
        
        elif event == 'TRANSITION A-D':
            logP += np.log(c)
            num_active -= 1
            num_dormant += 1
            if log:
                print("TRANSITION A-D: ", np.log(c))
        
        elif event == 'SAMPLE A':
            num_active += 1
        
        elif event == 'SAMPLE D':
            num_dormant -= 1
        
        else:
            raise Exception
        
        if log:
            print("NEW LOGP: ", logP)
            print()
        
    assert num_active == 1

    return logP
```

```python
def calculateDensity_(newick, c, K, N, log=False):
    with open("./temp.txt", 'w') as file:
        file.write(newick)

    tree = Phylo.read("./temp.txt", 'newick', rooted=True)
    os.remove("./temp.txt")
    
    return calculateDensity(tree, c, K, N, log)
```
